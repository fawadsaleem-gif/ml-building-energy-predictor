```python
import pickle
import pandas as pd
```

```python

#  Load saved model
loaded_model = pickle.load(open("finalised_model_SelectKReg.sav", "rb"))

# Enter Building Values
print("=" * 45)
print("   BUILDING ENERGY EFFICIENCY PREDICTOR")
print("=" * 45)

compactness  = float(input("Relative Compactness (0.62 - 0.98): "))
print("\nBuilding Type → 0=Low Rise, 1=Mid Rise, 2=High Rise")
building_type = int(input("Building Type: "))
surface      = float(input("Surface Area m²  (514.5 - 808.5): "))
wall         = float(input("Wall Area m²     (245.0 - 416.5): "))
roof         = float(input("Roof Area m²     (110.25 - 220.5): "))
height       = float(input("Overall Height m (3.5 or 7.0): "))
glazing      = float(input("Glazing Area Ratio (0.0 - 0.4): "))


new_building = pd.DataFrame([[
    compactness,
    building_type,
    surface,
    wall,
    roof,
    height,
    glazing
]], columns=[
    'Relative_Compactness',
    'Building_Type',
    'Surface_Area_m2',
    'Wall_Area_m2',
    'Roof_Area_m2',
    'Overall_Height_m',
    'Glazing_Area_ratio'
])

# ── Predict ───────────────────────────────────────────────────────
result = loaded_model.predict(new_building)

# ── Display Result ────────────────────────────────────────────────
print("\n--- RESULT ---")

if result[0] < 15:
    rating = "A — Very Efficient"
elif result[0] < 25:
    rating = "B — Efficient"
elif result[0] < 35:
    rating = "C — Moderate"
else:
    rating = "D — High Energy Use"

print(f"Predicted Heating Load : {result[0]:.2f} kWh/m²")
print(f"Energy Rating          : {rating}")
```
